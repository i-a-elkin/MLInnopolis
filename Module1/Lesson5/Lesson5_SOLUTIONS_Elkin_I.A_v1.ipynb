```python
from math import comb
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

## Задание 1
В ящике находится 15 качественных и 5 бракованных деталей. Наудачу извлекаются 2 детали. Найти вероятность того, что: 

а) обе детали будут качественными;  
б) одна деталь будет качественной, а одна – бракованной;  
в) обе детали бракованны.

```python
# а)
p1 = comb(15, 2) / comb(20, 2)
print(f"Вероятность, что обе детали будут качественными: {p1:.2f}")

# б)
p2 = comb(15, 1) * comb(5, 1) / comb(20, 2)
print(
    f"Вероятность, что одна деталь будет качественной, а одна – бракованной: {p2:.2f}"
)

# в)
p3 = comb(5, 2) / comb(20, 2)
print(f"Вероятность, что обе детали бракованны: {p3:.2f}")
```

## Задание 2
При рентгеновском обследовании вероятность обнаружить заболевание туберкулезом у больного туберкулезом равна 0.9, вероятность принять здорового человека за больного равна 0.01. Доля 
больных туберкулезом по отношению ко всему населению равна 0.001. Найти вероятность того, что человек здоров, если он был признан больным при обследовании. 

* Подсказка: использовать формулу Байеса, событие A – обследование показало наличие туберкулеза, В1 – человек здоров, В2 – человек болен.

```python
# вероятность, что человек болен
P_B2 = 0.001
# вероятность, что человек здоров
P_B1 = 1 - P_B2

# вероятность диагностировать заболевание, если человек болен
P_A_B2 = 0.9
# вероятность диагностировать заболевание, если человек здоров
P_A_B1 = 0.01

# вероятность, что у человека диагностировали заболевание
P_A = P_A_B1 * P_B1 + P_A_B2 * P_B2

# вероятность, что человек здоров, если у него диагностировали заболевание
P_B1_A = P_A_B1 * P_B1 / P_A
print(f"Вероятность, что человек здоров, если он был признан больным: {P_B1_A:.2f}")
```

## Задание 3
Известно, что случайная величина «время до наступления отказа исследуемых изделий» подчиняется логнормальному распределению (lognorm) с параметрами loc=0, scale=5, s=0.5. 

Требуется построить графики функции плотности и функции распределения, найти математическое ожидание, дисперсию и время безотказной работы с вероятностью 0.9 (квантиль уровня 0.1).

```python
# логнормальное распределение с нужными параметрами
lognorm_distribution = stats.lognorm(s=0.5, loc=0, scale=5)

# графики PDF и CDF
x = np.linspace(0, 20, 1000)
pdf = lognorm_distribution.pdf(x)
cdf = lognorm_distribution.cdf(x)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(x, pdf,)
ax[0].set_title("PDF")
ax[0].set_xlabel("Время до отказа")
ax[0].set_ylabel("Плотность вероятности")
ax[0].grid(True)

ax[1].plot(x, cdf, color="orange")
ax[1].set_title("CDF")
ax[1].set_xlabel("Время до отказа")
ax[1].set_ylabel("Накопленная вероятность")
ax[1].grid(True)

plt.tight_layout()
plt.show()

# параметры распределения
print(f"Математическое ожидание времени до отказа: {lognorm_distribution.mean():.2f}")
print(f"Дисперсия времени до отказа: {lognorm_distribution.var():.2f}")
print(f"Время безотказной работы с вероятностью 0.9 (квантиль уровня 0.1): {lognorm_distribution.ppf(0.1):.2f}")
```
